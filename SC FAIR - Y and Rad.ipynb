```python
#Here we will be plotting the y dimension along with each values radiation level
#This will also be bivariate with a KDE function to plot it
#EACH Y POINT MUST FOLLOW THE INEQUALITY:
Gamma (Y) > Gamma (background)
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'X': [15,30,30,30,20,15],
        'Y': [0,1,2,3,4,5],
    }
)

#NOTE THAT X HERE IS THE LEVEL OF Y IN CPM, AND Y IS THE Y DIMENSION
```

```python
sns.set_theme(style="white")

g = sns.JointGrid(data=df, x='X', y='Y', space=0);
g.plot_joint(sns.kdeplot,
             fill=True,
             levels=1000,
             cmap="mako",
             thresh=0
            
            );
g.plot_marginals(sns.kdeplot, fill=True);
```
