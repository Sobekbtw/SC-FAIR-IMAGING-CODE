```python
#Here we will be plotting the x dimension along with each values radiation level
#This will also be bivariate with a KDE function to plot it
#EACH X POINT MUST FOLLOW THE INEQUALITY:
Gamma (X) > Gamma (background)
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'X': [3,4,5,6,7],
        'Y': [15,30,30,30,15],
    }
)

#NOTE THAT X HERE IS THE X DIMENSION, AND Y IS THE RADIATION LEVEL OF X IN CPM
```

```python
sns.set_theme(style="white")

g = sns.JointGrid(data=df, x='X', y='Y', space=0);
g.plot_joint(sns.kdeplot,
             fill=True,
             levels=1000,
             cmap="mako",
             thresh=0
            
            );
g.plot_marginals(sns.kdeplot, fill=True);
```
