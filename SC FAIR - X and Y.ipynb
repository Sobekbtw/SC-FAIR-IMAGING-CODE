```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

df = pd.DataFrame(
    {
        'X': [3,5,5,6,7],
        'Y': [0,1,2,3,4],
    }
)
```

```python
sns.set_theme(style="white")

g = sns.JointGrid(data=df, x='X', y='Y', space=0);
g.plot_joint(sns.kdeplot,
             fill=True,
             levels=1000,
             cmap="rocket",
             thresh=0
            
            );
g.plot_marginals(sns.kdeplot, fill=True);
```
